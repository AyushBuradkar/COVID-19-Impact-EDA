```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
raw_data = pd.read_csv('raw_data.csv')

transformed_data = pd.read_csv('transformed_data.csv')
```

```python
print("Raw Data Info:")
print(raw_data.info())

print("\nTransformed Data Info:")
print(transformed_data.info())


print("\nRaw Data Summary Statistics:")
print(raw_data.describe())


print("\nTransformed Data Summary Statistics:")
print(transformed_data.describe())
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=raw_data, x='location', y='total_cases', ci=None, label='Total Cases', color='blue')
sns.barplot(data=raw_data, x='location', y='total_deaths', ci=None, label='Total Deaths', color='red')
plt.title('COVID-19 Cases and Deaths by Location (Raw Data)')
plt.xlabel('Location')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=raw_data, x='gdp_per_capita', y='total_cases')
plt.title('Scatter Plot: GDP per Capita vs Total Cases (Raw Data)')
plt.xlabel('GDP per Capita')
plt.ylabel('Total Cases')
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(transformed_data[['HDI', 'TC', 'TD', 'STI', 'POP', 'GDPCAP']])
plt.suptitle('Pairplot of Numerical Columns (Transformed Data)', y=1.02)
plt.tight_layout()
plt.show()
```
